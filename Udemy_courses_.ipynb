#price

```python
df['price'].unique()
#shows all values of price column
#found there are 2 string values (free and True)
#so that we can change them to INT with (0) value
```

```python
#changing all values to numeric
#getting the average of the price column
df['price']= df['price'].replace('Free', 0, regex=True)
df['price']= df['price'].replace('TRUE', 0, regex=True)
df['price'] = df['price'].astype(float)
average_price = df['price'].mean()
print(f"The average price is: {average_price}")
```

```python
#Getting the max and min of price column
maximum_price = df['price'].max()
print(f"The maximum price is: {maximum_price}")
print('__________________________________________________')
minimum_price = df['price'].min()
print(f"The minimum price is: {minimum_price}")
```

```python
#make the profit column
#profit=price*num_customers
df['profit']=df['price']*df['num_subscribers']
df.head()
```

```python
MAX=df['profit'].max()
print(f"The maximum profit is: {MAX}")
print('__________________________________________________')
MIN=df['profit'].min()
print(f"The minimum profit is: {MIN}")
print('__________________________________________________')
location_max=df['profit'].idxmax()
print(f"The location of the maximum profit is: {location_max}")
print('__________________________________________________')
df.loc[location_max]
```

```python
#changing the place of profit column
cols = list(df.columns)
profit_col = df.pop('profit')
price_index = cols.index('price')
df.insert(price_index - 1, 'profit', profit_col)
```

```python

plt.figure(figsize=(15, 7))
df_sorted_profit = df.sort_values('profit', ascending=False)
plt.bar(df_sorted_profit['subject'], df_sorted_profit['profit'])
plt.xlabel('Subject')
plt.ylabel('Profit')
plt.title('Profit by Subject')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
